### Qu es un DataFrame? 

Un dataframe puede ser interpretado como una tabla de datos, en la que tenemos una serie de filas y de columnas.

Cada **fila** representa un individuo o evento, por ejemplo un cliente de un banco, un producto de un supermercado o un coche que pasa por un peaje. 

En las **columnas** estn representadas las caractersticas de estos individuos o eventos: su procedencia, tamao o el da y la hora en que fue tomado el dato.

#### Calcular nuevas variables

Existen dos mtodos para introducirse en el corazn de un dataframe:

 - loc
 - iloc


Con **loc** entramos a travs de las variables/columnas, con **iloc** a travs de las filas/indices

**loc** es muy utilizado para crear nuevas variables

```python
# lgica de la sintaxis de un loc
# dataframe.loc[dataframe[variable] condicion, nombre_nueva_variable] = valor_que_toma_si_cumple_condicion

familia_df.loc[familia_df['edad'] >= 18, 'adulto'] = 'Si'
familia_df.loc[familia_df['edad'] < 18, 'adulto'] = 'No'
familia_df
```

Por otro lado **iloc** es usado para acceder a la tabla a travs de sus ndices

```python
# Accedemos a la informacin de la tabla a travs de la posicin de las filas y las columnas.
# dataframe.iloc[rango de filas que queremos retener , rango de columnas que queremos retener]

familia_df.iloc[1:3,2:5]
```

#### Agrupar datos

Podemos calcular operaciones matemticas en funcin de una variable categrica. Primeramente agrupamos las filas de la tabla inicial por la variable estado civil, posteriormente calculamos la media, la desviacin tpica y el valor mximo de las variables edad y altura

```python
# otras funciones que podemos calcular: min, count, median, sum
familia_df.groupby('estado_civil')[['edad','altura']].agg(['mean','std','max'])
```

#### Juntar tablas

Principalmente contamos con dos maneras para juntar informacin de dos tablas diferentes: 

 - por filas, axis = 0
 - por columnas, axis = 1
 
Por otro lado, tenemos dos mtodos que nos permiten hacer esta unin (con la librera pandas):

 - concat
 - merge
 
Primeramente vamos a crear otro dataframe, en este caso contendr informacin acerca de la situacin laboral de unas personas

```python
trabajadores = {'nombre':['Manuel','Maria','Diego','Bernarda','Alejandro','Felipe','Concha'],
           'antiguedad': [14, 26, 1, 2, 0, 6, 50],
           'profesion':['administrativo','director','becario','analista de datos','desempleado','abogado','retirada'],
           'salario':[1000, 2000, 100, 1000, 0, 1500, 800]}

trabajadores_df = pd.DataFrame(trabajadores)
trabajadores_df
```

Cuando juntamos por filas hay que tener en cuenta que ambas tablas tengan las mismas columnas, si no crearemos valores missing

Por otro lado, cuando juntamos por columnas es deseable que haya filas en comn, segn la columna de cruce

```python
# how puede tomar los siguientes valores: left, inner, right y outer
cruce_merge = familia_df.merge(trabajadores_df, on = 'nombre', how = 'left') 
cruce_merge
```

```python
cruce_concat0 = pd.concat([familia_df, trabajadores_df], axis = 0)
cruce_concat0
```

```python
cruce_concat1 = pd.concat([familia_df, trabajadores_df], axis = 1)
cruce_concat1
```

## Medidas de tendencia central y de dispersin

 - Media, suma de los valores dividido entre el nmero de elementos
 - Mediana, ordenamos los valores ascendentemente y nos quedamos con el valor de en medio
 - Moda, el valor ms repetido
 
 
 - Varianza, sumamos la distancia de cada elemento a la media y dividimos por el nmero de elementos
 - Desviacin tpica, raz cuadrada de la varianza
 
 
 - Ventajas y desventajas de la Media y la Mediana:
   - La media estamos mucho ms habituados a ella y es ms fcil de entender y calcular
   - La media se ve muy afectada por los valores extremos
   - La media solo debe ser usada cuando la distribucin de los datos sigue una distribucin normal
   - La mediana es ms robusta a valores extremos y no precisa de condiciones como seguir una distribucin normal
