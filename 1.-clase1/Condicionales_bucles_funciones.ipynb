## FOR

Los bucles FOR estn ideados para pasar por los elementos de un objeto. Normalmente son listas. 

Hay dos formas de pasar por estos **elementos**, a travs de cada uno de ellos o a travs de sus **ndices**

En este caso "viajamos" por los diferentes elementos de la lista numeros. En cada vuelta num toma el valor de un elemento, en la vuelta siguiente cambiar al prximo

```python
numeros = [1, 2, 3, 5, 7, 8]
for num in numeros:
    print("numero", num)
```

Los ndices son las posiciones en las que se encuentran cada uno de los elementos, como si fueran cajitas. En la primera vuelta vamos a la cajita 0 e imprimimos su elemento (index 0 numero 1). En las iteraciones siguientes aumenta en uno el ndice y con l la informacin que contiene (numero 2, numero 3...)

```python
numeros = [1, 2, 3, 5, 7, 8]
for i in range(len(numeros)):
    print("index", i, "numero", numeros[i])
```

Para saber si un bucle for est buscando la informacin de la lista por elemento o por ndice basta fijarse en la segunda parte de su definicin, a partir del "in", si tiene directamente la lista (como en el primer ejemplo "numeros") busca la informacin elemento por elemento, si en cambio tiene un rango estamos buscando por ndice, ya que le estamos indicando cuntas veces tiene que iterar, en nuestro caso 6, que es la longitud/tamao (len) de la lista numeros

a continuacin incluimos una condicin dentro del bucle FOR

```python
numeros = [1, 2, 3, 5, 7, 8]
for num in numeros:
    if num >= 3:
        print("numero", num)
    else:
        print("es muy bajo el numero para imprimir")
```

```python
numeros = [1, 2, 3, 5, 7, 8]
for i in range(len(numeros)):
    if i > 3:
        print("index", i, "numero", numeros[i])
```

## WHILE

Cuando iniciamos un bucle WHILE se repetirn todas las acciones que hay dentro de l hasta que se cumpla una condicin

 Atencin asegurarse de que esa condicin se cumple en algn momento!!

```python
count = 0
while count < 5:
    print(count)
    count += 1
    
```

```python
count = 0
while count < 5:
    if count % 2 != 0:
        print(count)
        count += 1
```

## BREAK

Este comando es usado cuando quieres parar el bucle, sea FOR o WHILE

```python
count = 0
while count < 5:
    if count % 2 != 0:
        print(count)
        break
    count += 1
```

## CONTINUE

CONTINUE es usado cuando quieres salir del bloque en el que te encuentras pero no parar la ejecucin o el bucle. En este caso sale de la condicin IF y vuelve al inicio del WHILE, sin pasar por el count += 1. Lo que ocurre aqu es que como no le suma 1 al objeto count siempre vale 0, por lo tanto siempre es menor que cinco y entonces nunca para de ejecutar, el WHILE no tiene fin.

```python
count = 0
while count < 5:
    if count % 2 != 0:
        print(count)
        continue
    count += 1
```

En este caso s tiene fin porque suma un valor a count antes del CONTINUE, sin embargo el segundo print("continue") nunca es ejecutado

```python
count = 0
while count < 5:
    count += 1
    if count % 2 != 0:
        print(count)
        continue
        print("continue")
    
```

## FUNCIONES

Herramientas que nosostros creamos a nuestro antojo para que realicen las tareas que ms nos convengan.

Tienen **argumentos** de entrada, que no siempre son obligatorios

El bucle FOR va iterando por cada uno de los key y los value del diccionario, segn la condicin imprime en pantalla una cosa u otra

```python
name1 = "Jonh"
birth_year1 = 1989

name2 = "Mari"
birth_year2 = 1995

name3 = "Mich"
birth_year3 = 1960
    
Dictionary={name1:birth_year1, name2:birth_year2, name3:birth_year3}
print(Dictionary)
```

```python
def generation_calculator(Dictionary):
    for key,value in Dictionary.items():
        if value <1949:
            print(key,"Silent Generation")
        elif value<1969:                  
            print(key,"Baby Boomer")
        elif value<1981:
            print(key,"X Generation")
        elif value<1994:
            print(key,"Millenials")
        else:
            print(key,"Z Generation") 
```

```python
generation_calculator(Dictionary)
```
