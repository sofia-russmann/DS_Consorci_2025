```python
df_fifa = df_fifa.apply(lambda col: col.map(lambda x: x.split('★')[0] if isinstance(x, str) and '★' in x else x))

print(df_fifa)
```

```python
df_fifa[['BP','Position']]
```

```python
mapeo_bp = {
    'CM': 'Centrocampista',
    'CDM': 'Centrocampista Defensivo',
    'ST': 'Delantero Centro',
    'LW': 'Extremo Izquierdo',
    'LM': 'Mediocampista Izquierdo',
    'CAM': 'Centrocampista Ofensivo',
    'CF': 'Delantero Centro',
    'LB': 'Lateral Izquierdo',
    'GK': 'Portero',
    'RM': 'Mediocampista Derecho',
    'RB': 'Lateral Derecho'
}

df_fifa['BP'] = df_fifa['BP'].map(mapeo_bp)
```

```python
df_fifa[['BP','Position']]
```

```python
df_fifa.Value
```

```python
def convertir_valor(valor):
    if 'M' in valor:
        return int(float(valor.replace('€', '').replace('M', '')) * 1_000_000)
    elif 'K' in valor:
        return int(float(valor.replace('€', '').replace('K', '')) * 1_000)
    else:
        return int(valor.replace('€', ''))

# Aplicar la función a la columna 'Value'
df_fifa['Value'] = df_fifa['Value'].apply(convertir_valor)
```

```python
df_fifa.Value
```

```python
df_fifa.Wage
```

```python
df_fifa['Wage'] = df_fifa['Wage'].apply(convertir_valor)
```

```python
df_fifa.Wage
```

```python
summary = df_fifa.groupby(['Nationality', 'Club', 'BP', 'Age']).agg({'Value': ['sum', 'mean'], 'Wage': ['sum', 'mean']})

print(summary)
```
